# Minimal test for FEMViewer

This notebook creates a simple mesh, wraps it in a FEMViewer, and tests display + update.

```python
import numpy as np

import FEmodel
import fileutils
import K3Dviewer

# Get data from the hole tensile test
disp_hole, force_hole, time_hole = fileutils.extract_data_from_tensile_test("Test Data/Hole/Tensile Data/data.csv", segment_number=1, tare=True)

# Resample the data to have 21 points in order to speed up the FE simulation
disp_hole_resampled = np.linspace(0, np.max(disp_hole), 21)
force_hole_resampled = np.interp(disp_hole_resampled, disp_hole, force_hole)

model = FEmodel.FEModel(75, 41, thickness=1, hole_diameter=7.2, ring_diameter=25.)
model.set_material_parameters_NH(C1_rubber=0.4, C1_ring=1.0)

view = K3Dviewer.MeshViewer(model, disp_hole_resampled)
view.display()
```

```python
forces, saved_u, saved_e = model.solve(disp_hole_resampled)
view.update_data(saved_u, saved_e, forces)
```

```python
# Get the experimental data from DIC
exp_fields = fileutils.read_mat_data("Test Data/Hole/DIC Data/Data/hole_rubber_last.mat")

# Plot the strain field in the undeformed configuration
FE_fields = model.get_fields_on_undeformed_mesh()
viewer = K3Dviewer.plot_fields_side_by_side(exp_fields, "exx", FE_fields, "axial_strain", "Axial Strain [-]")

display(viewer)
```
